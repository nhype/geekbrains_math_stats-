1.Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks): zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110], ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату (то есть, zp - признак), а за y - значения скорингового балла (то есть, ks - целевая переменная). Произвести расчет как с использованием intercept, так и без

```
x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

n = len(x)
#с интерсептом
#y = a + bx

b = (n*np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2)-np.sum(x)**2)
a = np.mean(y) - b*np.mean(x)
print(f"С интерсептом, вычисление не матричным методом: y={a}+{b}*x")


#матричный метод c интерсептом

x=x.reshape((n,1))
y=y.reshape((n,1))
x=np.hstack([np.ones((n,1)),x])

B = np.dot(np.dot(np.linalg.inv(x.T@x), x.T), y)
print(f"С интерсептом, вычисление матричным методом: y={B[0][0]}+{B[1][0]}*x")

# матричный метод без интерсепта

x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
x=x.reshape((n,1))
y=y.reshape((n,1))

B = np.dot(np.dot(np.linalg.inv(x.T@x), x.T), y)
print(f"Без интерсепта, вычисление матричным методом: y={B[0][0]}*x")
```

2.Посчитать коэффициент линейной регрессии при заработной плате (zp), используя градиентный спуск (без intercept).

```
x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
def mse_(B1, x, y):
  n = len(x)
  return np.sum((B1*x-y)**2)/n

alpha = 1e-7
def mse_p(B1,x,y):
  n = len(x)
  return 2*np.sum((B1*x-y)*x)/n
B = 0.1
for i in range(4000):
  B -= alpha*mse_p(B, x,y)
  error = mse_(B, x,y)
  if i%500 == 0:
    print(f"Iter {i} B = {B}, mse_error = {error}")
print(f"Градиентным спуском получаем y={B}*x")
```

3. В каких случаях для вычисления доверительных интервалов и проверки статистических гипотез используется таблица значений функции Лапласа, а в каких - таблица критических точек распределения Стьюдента?

Z-критерий используем, когда известно среднеквадратичное (стандартное) отклонение генеральной совокупности, критерий Стьюдента когда неизвестно среднеквадратичное отклонение у генеральной совокупности

*4. Произвести вычисления как в пункте 2, но с вычислением intercept. Учесть, что изменение коэффициентов должно производиться на каждом шаге одновременно (то есть изменение одного коэффициента не должно влиять на изменение другого во время одной итерации).

```
x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
def mse_(B0, B1, x, y):
  n = len(x)
  return np.sum((B1*x+B0-y)**2)/n

alpha = 1e-5

def mse_p_b1(B0, B1, x,y):
  n = len(x)
  return 2*np.sum((B1*x-y+B0)*x)/n

def mse_p_b0(B0, B1, x, y):
  n = len(x)
  return 2*np.sum(B1*x-y+B0)/n

B0 = 10
B1= 0.1
for i in range(2000000):
  change_b0 = alpha * mse_p_b0(B0,B1,x,y)
  change_b1 = alpha * mse_p_b1(B0,B1,x,y)

  B0 -= change_b0
  B1 -= change_b1
  error = mse_(B0, B1, x,y)
  if i%10000 == 0:
    print(f"Iter {i} B0 = {B0}, B1= {B1}, mse_error = {error}")
print(f"Градиентным спуском получаем y={B1}*x+{B0}")
```
